```python
from manim import *
config.media_width = "100%"
config.media_embed = True
```

```python
%%manim $render_as_image

config.media_embed = True

class Example(Scene):
    def construct(self):
        c = Circle(fill_color=WHITE, fill_opacity=1, stroke_width=0, radius=3)
        t1 = Triangle(fill_color=TEAL_E, fill_opacity=1, stroke_width=0)
        t2 = Triangle(fill_color=BLUE_E, fill_opacity=1, stroke_width=0).scale(2/3).align_to(t1, DOWN)
        t3 = Triangle(fill_color=WHITE, fill_opacity=1, stroke_width=0).scale(1/3).align_to(t1, DOWN)
        logo = VGroup(t1,t2,t3).scale(2)
        logo.rotate(PI, about_point=ORIGIN)

        self.add(c,t1,t2,t3)
        
        
```

```python
%%manim $render_as_image

config.media_embed = True

class Example(Scene):
    def construct(self):
        fig_kwargs = {"fill_opacity": 1}
        circle = Circle(color=WHITE,**fig_kwargs).scale(3)

        big_triangle = Triangle(color="#2FA673",**fig_kwargs).scale(2.4)
        big_triangle.rotate(PI,about_point=ORIGIN)

        mid_triangle = big_triangle.copy()
        mid_triangle.scale(2/3,about_edge=UP).set_color("#2A3C4E")

        small_triangle = big_triangle.copy()
        small_triangle.scale(1/3,about_edge=UP).set_color(WHITE)


        self.add(circle,big_triangle,mid_triangle,small_triangle)
        
        
```
