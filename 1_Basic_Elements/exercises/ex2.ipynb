```python
%%manim $render_as_image

config.media_embed = True

class Example(Scene):
    def construct(self):
        circle = Circle()
        circle.to_edge(UL) #same as to_corner
        self.add(circle)
        objs = [Square(), Star(), Triangle(), Circle()]
        pobj = circle
        for obj in objs:
            obj.next_to(pobj, RIGHT, aligned_edge=UP)
            self.add(obj)
            pobj = obj
```
