```python
%%manim $render_as_image

config.media_embed = True

class Example(Scene):
    def construct(self):
        s = Square(fill_color=RED, fill_opacity=0.7,stroke_width=0,side_length=4)
        c1 = Circle(radius=2).next_to(s, LEFT, buff=-2).match_style(s)
        c2 = Circle(radius=2).next_to(s, UP, buff=-2).match_style(s)
        vg = VGroup(s, c1, c2)
        vg.rotate(-PI/4, about_point=ORIGIN)
        self.add(vg)
        
        
```
